```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

mega_df = pd.read_csv('uniq2_table.csv')


bins = np.arange(0, 241, 30)

# split the time into intervals using pd.cut()
mega_df['time_bin'] = pd.cut(mega_df['Time'], bins=bins, right=False)

# group by these intervals and calculate the sum of all ones in peak_australia
#    do it across all cases (i.e., without splitting by 'data')
grouped_sum = mega_df.groupby('time_bin')['peak_australia'].sum()

# calculate the average value across all bins
mean_value = grouped_sum.mean()

threshold = mean_value * 1.15

x_positions = np.arange(len(grouped_sum))  # indices along the X-axis
y_values = grouped_sum.values             # the sums themselves

plt.bar(x_positions, y_values, color='skyblue', edgecolor='black')
plt.xticks(x_positions, grouped_sum.index, rotation=45)  # X-axis labels — intervals

plt.xlabel('30 s')
plt.ylabel('count of peaks')
plt.title('peak_australia')

# add a black dot if the bin sum > threshold
for x, y in zip(x_positions, y_values):
    if y > threshold:
        plt.scatter(x, y, color='black', zorder=3)  # Отметка чёрной точкой

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



mega_df = pd.read_csv('3sec_table.csv')


bins = np.arange(0, 241, 5)


mega_df['time_bin'] = pd.cut(mega_df['Time'], bins=bins, right=False)


grouped_sum = mega_df.groupby('time_bin')['peak_germany'].sum()


mean_value = grouped_sum.mean()

threshold = mean_value * 2

x_positions = np.arange(len(grouped_sum))  # Индексы по осям X
y_values = grouped_sum.values             # Сами суммы

plt.bar(x_positions, y_values, color='skyblue', edgecolor='black')


plt.xlabel('5 s')
plt.ylabel('count of peaks')
plt.title('peak_germany')


for x, y in zip(x_positions, y_values):
    if y > threshold:
        plt.scatter(x, y, color='black', zorder=3)  

plt.tight_layout()
plt.show()
```
